## Para analizar las crisis economicas
- Para las crisis económicas decidimos seleccionar las 4 más severas durante el período de estudio, seleccionamos el año en el que comienza cada crisis y se contrasta con el año inmediatamente anterior a ese, sin irregularidades económicas.
- 1973: crisis del petróleo. La OPEP decide no exportar petróleo a los países que apoyaron a Israel en la guerra del Yom Kipur, esto provocó que se cuadruplicara el precio del petroleo debido a la escasez del bien. 
- 1979: segunda crisis del petróleo. Debido a la revolución iraní por el descontento con el sah y el [viernes negro](https://historia.nationalgeographic.com.es/foto-del-dia/viernes-negro-irani_16408), los trabajadores de las refinerias iranies se declararon en huelga, provocando de nuevo un alza en el precio de petroleo de 2.7 veces
- 1990: provocada por el shock del petróleo, el final de la guerra fría, crisis de ahorro y préstamos, y una caída en la construccion por sobreconstruccion en los años 80. 
-2007:Causada por el colapso del mercado inmobiliario en estados unidos, y la crisis de las hipotecas de alto riesgo. 
- Seleccionamos el año anterior a la crisis para analizar el cambio, y la disminución esperada de la producción, comparando un año afectado por irregularidades económicas y uno más estable.


```python
niño=[1973,1979,1982,1986,1987,1991,1994,1997,2002,2004,2006,2009,1983,1984,1988,1995,1998,1999,2000,2007,2010,2011]
econom=[1972,1973,1978,1979,1989,1990,2006,2007]
econo=[]
crisis_cli=[]
for año in dt["Year"]:
    if año in niño:
        crisis_cli.append("si")
    else:
        crisis_cli.append("no")
dt["Crisis Cli"]=crisis_cli
for año in dt["Year"]:
    if año in econom:
        econo.append("si")
    else:
        econo.append("no")
dt["Crisis Eco"]=econo
dt.head()

```

## Determinamos si hay diferencias significativas en la producción según crisis climática y económica considerando si se esta dentro o no de la G7


```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
dt = dt.rename(columns={"Crisis Eco":"Crisis_Eco"})
dt1_lm = ols("Production ~ C(Crisis_Eco, Sum)*C(G_7, Sum)",
             data=dt).fit()
table = sm.stats.anova_lm(dt1_lm, typ=2)
print(table)
dt = dt.rename(columns={"Crisis Cli":"Crisis_Cli"})
dt2_lm = ols("Production ~ C(Crisis_Cli, Sum)*C(G_7, Sum)",
             data=dt).fit()
table = sm.stats.anova_lm(dt2_lm, typ=2) 
print(table)
# https://www.statsmodels.org/stable/anova.html
```

## Se calcula el intercepto y pendiente de la regresión lineal y se agregan como columna
- En el caso de USA e INDI por tener tendencias propias se calculan estos valores de manera personalizada
- [referencia regresión 1](https://rpubs.com/joser/RegresionSimple)
- [referencia regresión 2](https://www.iartificial.net/regresion-lineal-con-ejemplos-en-python/)



```python
from sklearn.linear_model import LinearRegression
USA=dt[dt["Area Abbreviation"]=="USA"]
INDI=dt[dt["Area Abbreviation"]=="INDI"]
modelo1 = LinearRegression()
modelo1.fit(X = dt["Year"].values.reshape(-1, 1), y = dt["Production"].values)

modelo2 = LinearRegression()
modelo2.fit(X = USA["Year"].values.reshape(-1, 1), y =USA["Production"].values)

modelo3 = LinearRegression()
modelo3.fit(X = INDI["Year"].values.reshape(-1, 1), y = INDI["Production"].values)

ki=[modelo1.intercept_,modelo1.coef_[0],modelo2.intercept_,modelo2.coef_[0],modelo3.intercept_,modelo3.coef_[0]]
pivote["Coef"]=[ki[1],ki[1],ki[1],ki[1],ki[1],ki[1],ki[1],ki[5],ki[1],ki[1],ki[1],ki[1],ki[1],ki[1],ki[1],ki[1],ki[1],ki[1],ki[3]]
pivote["Inter"]=[ki[0],ki[0],ki[0],ki[0],ki[0],ki[0],ki[0],ki[4],ki[0],ki[0],ki[0],ki[0],ki[0],ki[0],ki[0],ki[0],ki[0],ki[0],ki[2]]
pivote
```

## Definimos una función para generar una columna por cada comparación de cada una de las 4 crisis económicas y extraer la proporción entre lo que cada país produjo en la realidad y lo que según la regresion lineal debía producir, aun contemplando la crisis económica.
## Si la proporción en cada columna es 1 o mayor, el país produjo lo que se esperaba en la crisis, si es menor a 1 no alcanzó la producción de comida esperada 


```python
def PC(x,y,coef,inter,a1,a2):
    d=(inter + coef*a2)-(inter + coef*a1)
    pc=(y-x)/d
    return(pc)
```

## Agregamos las columnas a la tabla pivote que se trabajó anteriormente 


```python
pivote["1972-1973"]=PC(pivote["Y1972"].values,pivote["Y1973"],pivote["Coef"].values,pivote["Inter"].values,1972,1973)
pivote["1978-1979"]=PC(pivote["Y1978"].values,pivote["Y1979"],pivote["Coef"].values,pivote["Inter"].values,1978,1979)
pivote["1989-1990"]=PC(pivote["Y1989"].values,pivote["Y1990"],pivote["Coef"].values,pivote["Inter"].values,1989,1990)
pivote["2006-2007"]=PC(pivote["Y2006"].values,pivote["Y2007"],pivote["Coef"].values,pivote["Inter"].values,2006,2007)
pivote
```
