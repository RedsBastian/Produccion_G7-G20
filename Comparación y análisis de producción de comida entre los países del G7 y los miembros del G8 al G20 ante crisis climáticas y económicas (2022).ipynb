## Determinamos si hay diferencias significativas en la producci칩n seg칰n crisis clim치tica y econ칩mica considerando si se esta dentro o no de la G7


```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
dt = dt.rename(columns={"Crisis Eco":"Crisis_Eco"})
dt1_lm = ols("Production ~ C(Crisis_Eco, Sum)*C(G_7, Sum)",
             data=dt).fit()
table = sm.stats.anova_lm(dt1_lm, typ=2)
print(table)
dt = dt.rename(columns={"Crisis Cli":"Crisis_Cli"})
dt2_lm = ols("Production ~ C(Crisis_Cli, Sum)*C(G_7, Sum)",
             data=dt).fit()
table = sm.stats.anova_lm(dt2_lm, typ=2) 
print(table)
# https://www.statsmodels.org/stable/anova.html
```

## Definimos una funci칩n para generar una columna por cada comparaci칩n de cada una de las 4 crisis econ칩micas y extraer la proporci칩n entre lo que cada pa칤s produjo en la realidad y lo que seg칰n la regresion lineal deb칤a producir, aun contemplando la crisis econ칩mica.
## Si la proporci칩n en cada columna es 1 o mayor, el pa칤s produjo lo que se esperaba en la crisis, si es menor a 1 no alcanz칩 la producci칩n de comida esperada 


```python
def PC(x,y,coef,inter,a1,a2):
    d=(inter + coef*a2)-(inter + coef*a1)
    pc=(y-x)/d
    return(pc)
```

## Agregamos las columnas a la tabla pivote que se trabaj칩 anteriormente 


```python
pivote["1972-1973"]=PC(pivote["Y1972"].values,pivote["Y1973"],pivote["Coef"].values,pivote["Inter"].values,1972,1973)
pivote["1978-1979"]=PC(pivote["Y1978"].values,pivote["Y1979"],pivote["Coef"].values,pivote["Inter"].values,1978,1979)
pivote["1989-1990"]=PC(pivote["Y1989"].values,pivote["Y1990"],pivote["Coef"].values,pivote["Inter"].values,1989,1990)
pivote["2006-2007"]=PC(pivote["Y2006"].values,pivote["Y2007"],pivote["Coef"].values,pivote["Inter"].values,2006,2007)
pivote
```
